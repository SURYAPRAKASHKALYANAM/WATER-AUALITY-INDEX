```python
import pandas as pd
df=pd.read_csv(r"Results_MADE.csv")
df
```

```python
df["Temperature"]=df["Temperature"].apply(lambda x:int(x))
```

```python
df["Dissolved Oxygen"]=df["Dissolved Oxygen"].apply(lambda x:int(x))
```

```python
df["Temperature"]=df["Temperature"].replace(0,df["Temperature"].mean())
```

```python
df["Dissolved Oxygen"]=df["Dissolved Oxygen"].replace(0,df["Dissolved Oxygen"].mean())
```

```python
df["Bio-Chemical Oxygen Demand (mg/L)"]=df["Bio-Chemical Oxygen Demand (mg/L)"].replace(0,df["Bio-Chemical Oxygen Demand (mg/L)"].mean())
```

```python
df["Faecal Streptococci (MPN/ 100 mL)"]=df["Faecal Streptococci (MPN/ 100 mL)"].replace(0,df["Faecal Streptococci (MPN/ 100 mL)"].mean())
```

```python
df["Nitrate (mg/ L)"]=df["Nitrate (mg/ L)"].replace(0,df["Nitrate (mg/ L)"].mean())
```

```python
df["Faecal Coliform (MPN/ 100 mL)"]=df["Faecal Coliform (MPN/ 100 mL)"].replace(0,df["Faecal Coliform (MPN/ 100 mL)"].mean())
```

```python
df["Total Coliform (MPN/ 100 mL)"]=df["Total Coliform (MPN/ 100 mL)"].replace(0,df["Total Coliform (MPN/ 100 mL)"].mean())
```

```python
df["Conductivity (mho/ Cm)"]=df["Conductivity (mho/ Cm)"].replace(0,df["Conductivity (mho/ Cm)"].mean())
```

```python
df.describe()
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=45)
```

```python
# # from sklearn.ensemble import RandomForestClassifier
# # from sklearn.svm import SVC
# from sklearn.naive_bayes import GaussianNB
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
# model=RandomForestClassifier()
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_test,y_pred))
```
